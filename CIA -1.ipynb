```python
import random
import time

def randstring(l):
    arr = ['A','T','C','G']
    res = ""
    for i in range(l):
        res += random.choice(arr)
    return res

def print_mat(mat, s1, s2):
    for i in range(len(s1)+1):
        for j in range(len(s2)+1):
            print(mat[i][j], end=' ')
        print()
    print()

def fill_array(mat, s1, s2, i=1, j=1):
    m, n = len(s1)+1, len(s2)+1
    global maxr, maxc

    if s1[i-1] == s2[j-1]:
        mat[i][j] = mat[i-1][j-1] + 2
    else:
        mat[i][j] = max(mat[i-1][j-1], mat[i][j-1], mat[i-1][j]) - 1

    if mat[i][j] > mat[maxr][maxc]:
        maxr, maxc = i, j

    if j < n-1:
        fill_array(mat, s1, s2, i, j+1)
    elif j == n-1 and i < m-1:
        fill_array(mat, s1, s2, i+1, 1)


maxr, maxc = 0, 0
def backtrack(mat, s1, s2, ans1="", ans2=""):
    n = len(s2)+1
    global maxr, maxc

    if maxr == 0 and maxc == 0:
        print(ans1)
        print(ans2)
        return
    elif mat[maxr-1][maxc-1] + 2 == mat[maxr][maxc] and s2[maxc-1] == s1[maxr-1]:
        ans2 = s2[maxc-1] + ans2
        ans1 = s1[maxr-1] + ans1
        maxr -= 1
        maxc -= 1
    elif mat[maxr][maxc-1] - 1 == mat[maxr][maxc]:
        ans2 = s2[maxc-1] + ans2
        ans1 = "_" + ans1
        maxc -= 1
    elif mat[maxr-1][maxc] - 1 == mat[maxr][maxc]:
        ans2 = "_" + ans2
        ans1 = s1[maxr-1] + ans1
        maxr -= 1
    else:
        ans2 = "*" + s2[maxc-1] + ans2
        ans1 = "*" + s1[maxr-1] + ans1
        maxr -= 1
        maxc -= 1

    backtrack(mat, s1, s2, ans1, ans2)

if __name__ == '__main__':
    s1 = "AGCACACA"
    s2 = "ACACACTA"
    print(s1)
    print(s2)

    mat = [[0 for j in range(len(s2)+1)] for i in range(len(s1)+1)]

    start = time.time()
    fill_array(mat, s1, s2)
    stop = time.time()

    print_mat(mat, s1, s2)
    print("time taken:", stop-start)

    backtrack(mat, s1, s2)
```
