```python
class PSO():
    def __init__(self, n_particles, dimensions, c1, c2, w):
        self.n_particles = n_particles
        self.dimensions = dimensions
        self.c1 = c1
        self.c2 = c2
        self.w = w
        self.particles = np.random.uniform(-1, 1, (n_particles, dimensions))
        self.velocities = np.zeros((n_particles, dimensions))
        self.best_positions = self.particles.copy()
        self.best_fitness = np.zeros(n_particles)
        
    def update(self, fitness):
        for i in range(self.n_particles):
            if fitness[i] > self.best_fitness[i]:
                self.best_positions[i] = self.particles[i].copy()
                self.best_fitness[i] = fitness[i]

        r1 = np.random.uniform(0, 1, (self.n_particles, self.dimensions))
        r2 = np.random.uniform(0, 1, (self.n_particles, self.dimensions))

        self.velocities = self.w * self.velocities \
                          + self.c1 * r1 * (self.best_positions - self.particles) \
                          + self.c2 * r2 * (self.best_positions.mean(axis=0) - self.particles)

        self.particles += self.velocities
```
